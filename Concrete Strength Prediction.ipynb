## Splitting the dataset

```python
# Matrix of features / independent variables
x = dataset.drop(columns='Concrete compressive strength(MPa, megapascals) ')
x.head()
```

```python
y = dataset['Concrete compressive strength(MPa, megapascals) ']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Part 2 - Building the model

## 1) Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor_lr = LinearRegression()
regressor_lr.fit(x_train,y_train)
```

```python
y_pred = regressor_lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

## 2) Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
regressor_rf = RandomForestRegressor()
regressor_rf.fit(x_train,y_train)
```

```python
y_pred = regressor_rf.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

## 3) XGBoost Regression

```python
from xgboost import XGBRFRegressor
regressor_xgb = XGBRFRegressor()
regressor_xgb.fit(x_train,y_train)
```

```python
y_pred = regressor_xgb.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

# Part 3 - Choosing the best model

We choose Random Forest Regression as it has the better r2 score of the other methods.
